```python
data = Path('heart_clean.csv')
heart_df = pd.read_csv(data)
heart_df.head()
```

```python
dup_heart_df = heart_df.drop_duplicates(subset=['age','sex',"resting_blood_pressure","serum_cholesterol"])
dup_heart_df.head(25)
```

```python
#create unique id column of patients
dup_heart_df['id'] = dup_heart_df[['age', 'sex','resting_blood_pressure','serum_cholesterol', 'fasting_blood_sugar','max_heart_rate_achieved']].sum(axis=1).map(hash)
dup_heart_df
```

```python
#sort by unique ID
final_heart_sorted_df = dup_heart_df.sort_values(by=['id'])
final_heart_sorted_df.head(25)
```

```python
final_heart_sorted_df = final_heart_sorted_df.drop_duplicates(subset=['id'])
final_heart_sorted_df
```

```python
output_file_path = "duplicates_removed_data.csv"
final_heart_sorted_df.to_csv(output_file_path, index=False)
```
